## Presque pas d'impact (par T1) sur le nombre de voitures sur tout Nice entre 2006 et 2016 

```python
g_p_2006_2016_Nice['TE_voit']=g_p_2006_2016_Nice['voit_2016']-g_p_2006_2016_Nice['voit_2006']
g_p_2006_2016_Nice['TE_voit'].hist()
```

```python
g_p_2006_2016_Nice['TE_voit'].mean()
```

```python
# voiture

fig, axs = plt.subplots(2,2, figsize=(15,10), squeeze=True)

max_min_2016=g_p_2006_2016_T1['voit_2016'].max()-g_p_2006_2016_T1['voit_2016'].min()
max_min_2006=g_p_2006_2016_T1['voit_2006'].max()-g_p_2006_2016_T1['voit_2006'].min()
    
    
axs[0][0].hist(g_p_2006_2016_T1['voit_2006'], bins=20)
axs[1][0].hist(g_p_2006_2016_T1['voit_2016'], bins=int((max_min_2016)*20/(max_min_2006)))
axs[0][0].set_title('nombre de voitures sur T1 en 2006')
axs[1][0].set_title('nombre de voitures sur T1 en 2016')
                                
max_min_2016=g_p_2006_2016_Nice['voit_2016'].max()-g_p_2006_2016_Nice['voit_2016'].min()
max_min_2006=g_p_2006_2016_Nice['voit_2006'].max()-g_p_2006_2016_Nice['voit_2006'].min()                                

axs[0][1].hist(g_p_2006_2016_Nice['voit_2006'], bins=20)
axs[1][1].hist(g_p_2006_2016_Nice['voit_2016'], bins=int((max_min_2016)*20/(max_min_2006)))
axs[0][1].set_title('nombre de voitures sur tout Nice en 2006')
axs[1][1].set_title('nombre de voiture sur tout Nice en 2016')
```
