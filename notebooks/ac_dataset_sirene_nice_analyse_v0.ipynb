### Classe de l'établissement

```python
pd.DataFrame(entreprise_ml_df["Classe de l'établissement"].unique())
```

```python
pd.DataFrame(entreprise_ml_df["Classe de l'établissement"].value_counts())
```

```python
ohe_etablissement = OneHotEncoder(sparse = False) # Instanciate encoder
ohe_etablissement.fit(entreprise_ml_df[["Classe de l'établissement"]]) # Fit encoder
```

```python
ohe_etablissement.categories_
```

```python
list(ohe_etablissement.categories_[0])
```

```python
etablissement_encoder=ohe_etablissement.transform(entreprise_ml_df[["Classe de l'établissement"]])
etablissement_encoder
```

```python
etablissement_encoder.shape
```

```python
pd.DataFrame(etablissement_encoder).head()
```

```python
add_df=pd.concat([
            entreprise_ml_df["Classe de l'établissement"],
            pd.DataFrame(etablissement_encoder,columns = list(ohe_etablissement.categories_[0]))
           ],axis=1)
add_df
```

```python
entreprise_ml_df_encoder_0=pd.concat([
                                entreprise_ml_df,
                                add_df    
                                ],axis=1)
entreprise_ml_df_encoder_0=entreprise_ml_df_encoder_0.drop(columns="Classe de l'établissement")
entreprise_ml_df_encoder_0
```
