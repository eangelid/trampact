```python
ohe_jurique = OneHotEncoder(sparse = False) # Instanciate encoder
ohe_jurique.fit(entreprise_df[["Nature juridique de l'unité légale"]]) # Fit encoder
```

```python
ohe_jurique.categories_
```

```python
list(ohe_jurique.categories_[0])
```

```python
jurique_encoder=ohe_jurique.transform(entreprise_df[["Nature juridique de l'unité légale"]])
jurique_encoder
```

```python
jurique_encoder.shape
```

```python
pd.DataFrame(jurique_encoder).head()
```

```python
add_df=pd.concat([
            entreprise_df["Nature juridique de l'unité légale"],
            pd.DataFrame(jurique_encoder,columns = list(ohe_jurique.categories_[0]))
           ],axis=1)
add_df
```

```python
entreprise_df_encoder=pd.concat([
                                entreprise_df,
                                add_df    
                                ],axis=1)
entreprise_df_encoder=entreprise_df_encoder.drop(columns="Nature juridique de l'unité légale")
entreprise_df_encoder
```
